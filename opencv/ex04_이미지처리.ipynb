# 이미지 처리 및 ROI

### 픽셀 기반 처리
- 이미지는 픽셀이라는 데이터 유닛의 집합
- R, G, B 세 가지 채널을 가지고 있음(png 데이터는 알파까지 4채널)
- c * h * w의 매트릭스

### 블록 기반 처리
- 블록 : 픽셀을 그룹화한 것

### 주파수 기반 처리
- 이미지에서의 주파수 : 이미지 픽셀값의 변화
- 고주파 : 픽셀값의 변화가 큼
- 저주파 : 픽셀값의 변화가 작음

```python
# 실습1 - 이미지 픽셀값 출력
import cv2

image = cv2.imread('images/01_lenna.png')
print(image)
```

```python
print(image[200,100])
```

```python
# 실습2 - 픽셀값 변경
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('images/01_lenna.png')
image[200, :, :] = [0, 0, 0]
b, g, r = cv2.split(image)
new_img = cv2.merge([r, g, b])
plt.imshow(new_img)
plt.show()
```

### 블록 기반 처리 ROI (Region Of Interest)

```python
import cv2

image = cv2.imread('images/01_lenna.png')
subimg = image[35:300, 150:350]
b, g, r = cv2.split(subimg)
new_img = cv2.merge([r, g, b])
plt.imshow(new_img)
plt.show()
```

### 도형 그리기

- cv2.elipse : 타원
